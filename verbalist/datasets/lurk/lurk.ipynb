```python
from datasets import load_dataset, Dataset, concatenate_datasets

dataset = load_dataset("averoo/lurk")
dataset = dataset["train"]
dataset
```

```python
dataset[100]
```

```python
new_dataset = []

for item in dataset:
    word = item['header']
    definition = item['text']
    prompt = f"напиши определение для {word} в стиле lurk"
    new_dataset.append({
		'prompt': prompt,
		"text": definition,
		"word": word
	})
```

```python
new_dataset = Dataset.from_list(new_dataset)
new_dataset.push_to_hub("dim/lurk_prompts")
```
